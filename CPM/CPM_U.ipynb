```python
#### contains a single line ####
line = list() 
Ele = list()

##### contains all the tks #####
tks = dict() 
number = -1
fhand = open('C:\\Users\\USER\\Desktop\\Class\\CPM\\cpm1.txt')

#### slide the file line by line ####
for line in fhand: 
    Ele=(line.split(',')) 
    number += 1

	#### creating the single task element ####
    for i in range(len(Ele)): 
        tks[str(Ele[0])]= dict()
        tks[str(Ele[0])]['id'] = Ele[0]
        tks[str(Ele[0])]['name'] = Ele[1]
        tks[str(Ele[0])]['time'] = Ele[2]
        if(Ele[3] != "\n"):
            tks[str(Ele[0])]['depend'] = Ele[3].strip().split(';')
        else:
            tks[str(Ele[0])]['depend'] = ['-1']
        tks[str(Ele[0])]['ES'] = 0
        tks[str(Ele[0])]['EF'] = 0
        tks[str(Ele[0])]['LS'] = 0
        tks[str(Ele[0])]['LF'] = 0
        tks[str(Ele[0])]['ST'] = 0
        tks[str(Ele[0])]['isCritical'] = False

# =============================================================================
# FORWARD PASS
# =============================================================================

#### Slides all the tks ####
for taskFW in tks: 
    #### checks if it's the first task ####
    if('-1' in tks[taskFW]['depend']): 
        tks[taskFW]['ES'] = 0
        tks[taskFW]['EF'] = (tks[taskFW]['time'])
    else: #### not the first task ####
        for k in tks.keys():
            #### slides all the dependency in a single task ####
            for dep in tks[k]['depend']: 
                #### if the task k has only one dependency ####
                if(dep != '-1' and len(tks[k]['depend']) == 1): 
                    tks[k]['ES'] = int(tks[ dep]['EF']) 
                    tks[k]['EF'] = int(tks[k]['ES']) + int(tks[k]['time']) 
                elif(dep !='-1'): 
                #### put the maximum value of ES for the depend of a task ####
                    if(int(tks[dep]['EF']) > int(tks[k]['ES'])):
                        tks[k]['ES'] = int(tks[ dep]['EF']) 
                        tks[k]['EF'] = int(tks[k]['ES']) + int(tks[k]['time']) 

aList = list() #list of task keys
for element in tks.keys():
    aList.append(element)

bList = list() #reversed list of task keys
while len(aList) > 0:
    bList.append(aList.pop())
    
# =============================================================================
# BACKWARD PASS
# =============================================================================
for taskBW in bList:
    #### check if it's the last task ####
    if(bList.index(taskBW) == 0):  
        tks[taskBW]['LF']=tks[taskBW]['EF']
        tks[taskBW]['LS']=tks[taskBW]['ES']
        
    for dep in tks[taskBW]['depend']: #slides all the dependency in a single task
        #### check if it's NOT the last task ####
        if(dep != '-1'): 
            #### check if the the dependency is already analyzed ####
            if(tks[dep]['LF'] == 0): 
                tks[dep]['LF'] = int(tks[taskBW]['LS']) 
                tks[dep]['LS'] = int(tks[dep]['LF']) - int(tks[dep]['time']) 
                tks[dep]['ST'] = int(tks[dep]['LF']) - int(tks[dep]['EF'])
            #### put the minimun value of LF for thedepend of a task ####
            if(int(tks[dep]['LF']) > int(tks[taskBW]['LS']) ): 
                tks[dep]['LF'] = int(tks[taskBW]['LS']) 
                tks[dep]['LS'] = int(tks[dep]['LF']) - int(tks[dep]['time']) 
                tks[dep]['ST'] = int(tks[dep]['LF']) - int(tks[dep]['EF'])
                
# =============================================================================
# PRINTING  
# =============================================================================
print('task id, task name, time, ES, EF, LS, LF, ST, isCritical')
for task in tks:
    if(tks[task]['ST'] == 0):
        tks[task]['isCritical'] = True
    print(str(tks[task]['id']) +', '+str(tks[task]['name']) +', '\
			+str(tks[task]['time']) +', '+str(tks[task]['ES']) +', '\
			+str(tks[task]['EF']) +', '+str(tks[task]['LS']) +', '+\
			str(tks[task]['LF']) +', '+str(tks[task]['ST']) +', '+\
			str(tks[task]['isCritical']))









```
